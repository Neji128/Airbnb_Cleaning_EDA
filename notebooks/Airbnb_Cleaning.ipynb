```python
def data_inspector(df):
    
       
    duplicates = df[df.duplicated()]
        
    display('----------------Missing Values----------------',df.isna().sum())
    display('----------------Duplicate Values----------------')
    display(str(len(duplicates))  + ' duplicates')
```
